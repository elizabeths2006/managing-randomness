```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def plot_ratio_data(DataSet, file1, file2, file3, file4, file5, file6, file7, file8, file9):
    """Plot the final accuracies of each model vs the number of clusters used in the making of the models"""
    #Set up plot
    x_axis = [2, 3, 4, 5, 6]  # set x-axis to cluster numbers
    plt.figure(figsize=(12, 8), dpi=80)
    plt.xlabel("# of Clusters")
    plt.xticks(np.arange(1, 7, 1))
    plt.ylabel("Final accuracy")
    plt.ylim([0,1])
    plt.yticks(np.arange(0, 1.1, .1))

    # Plot both silhouette scores and standard dev data
    plot1 = plt.plot(x_axis, np.loadtxt(file1, dtype=float), "-o", label='10/90')
    plot2 = plt.plot(x_axis, np.loadtxt(file2, dtype=float), "-o", label='20/80')
    plot3 = plt.plot(x_axis, np.loadtxt(file3, dtype=float), "-o", label='30/70')
    plot4 = plt.plot(x_axis, np.loadtxt(file4, dtype=float), "-o", label='40/60')
    plot5 = plt.plot(x_axis, np.loadtxt(file5, dtype=float), "-o", label='50/50')
    plot6 = plt.plot(x_axis, np.loadtxt(file6, dtype=float), "-o", label='60/40')
    plot7 = plt.plot(x_axis, np.loadtxt(file7, dtype=float), "-o", label='70/30')
    plot8 = plt.plot(x_axis, np.loadtxt(file8, dtype=float), "-o", label='80/20')
    plot9 = plt.plot(x_axis, np.loadtxt(file9, dtype=float), "-o", label='90/10')


    # Add plot legend and title
    plt.legend()
    plt.title("Varied Clusters and Train/Test Ratio - " + DataSet + " (K-Means)")
    plt.show()
```

# Iris

```python
## Iris data set

plot_ratio_data("Iris", "iris-data/iris_ratio1_accuracy.txt",  "iris-data/iris_ratio2_accuracy.txt",
                "iris-data/iris_ratio3_accuracy.txt",  "iris-data/iris_ratio4_accuracy.txt",
                "iris-data/iris_ratio5_accuracy.txt",  "iris-data/iris_ratio6_accuracy.txt",
                "iris-data/iris_ratio7_accuracy.txt",  "iris-data/iris_ratio8_accuracy.txt",
                "iris-data/iris_ratio9_accuracy.txt")
```
