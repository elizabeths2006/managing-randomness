```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import genfromtxt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
def calcSilhScores(trainSet):
    """Plot and return the silhouette scores versus the number of possible clusters for the given dataset"""
    scores_dict = {}
    range_n_clusters = [2, 3, 4, 5, 6]
    for n_clusters in range_n_clusters:

        # Instantiate the KMeans models
        km = KMeans(n_clusters=n_clusters)

        # Fit the KMeans model
        km.fit_predict(trainSet)

        # Calculate Silhoutte Score
        score = silhouette_score(trainSet, km.labels_, metric='euclidean')

        # Add score to dictionary
        scores_dict[n_clusters] = score
        
    return scores_dict
        
def plotSilhScores(DataSetName, trainSet):
    """Plot silhouette scores for a data set"""
    scores_dict = calcSilhScores(trainSet) # Get silhouette scores
    
    # Plot silhouette scores
    x_axis = scores_dict.keys()
    y_axis = scores_dict.values()
    plt.figure(figsize=(8, 6), dpi=80)
    plt.scatter(x_axis,y_axis)

    # Set x and y axis intervals
    plt.xlim([1,7])
    plt.xticks(np.arange(1,7,1))

    plt.ylim([0,1])
    plt.yticks(np.arange(0, 1, .1))
    
    # Plot average line
    plt.axhline(y=np.mean(list(y_axis)), color='r', linestyle='-', label="average plot")


    # Label, title, and show plot
    plt.xlabel("# of Clusters")
    plt.ylabel("Silhouette Score")
    plt.title("Silhouette Scores vs # of Clusters - " + DataSetName + " (K-Means)")
    plt.legend()
    plt.show()
```

# Iris

```python
## Iris data set ##

# Load IRIS dataset from sklearn
iris = datasets.load_iris()
X = iris.data
y = iris.target

iris_from_txt = np.loadtxt("iris-data.txt")

# Plot scores and model performances
print("Using 30% as training data:")
plotSilhScores("Iris", X)

print("Using whole data set as training data:")
plotSilhScores("Iris",iris_from_txt)
```
