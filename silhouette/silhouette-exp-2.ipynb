```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import genfromtxt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
def plotAccVsClusters(DataSetName, file_clusters2, file_clusters3, file_clusters4, file_clusters5, file_clusters6):
    """Plot the final accuracies of each model vs the number of clusters used in the making of the models"""
    # Load data as numpy arrays
    clusters2 = np.loadtxt(file_clusters2, dtype=float)
    clusters3 = np.loadtxt(file_clusters3, dtype=float)
    clusters4 = np.loadtxt(file_clusters4, dtype=float)
    clusters5 = np.loadtxt(file_clusters5, dtype=float)
    clusters6 = np.loadtxt(file_clusters5, dtype=float)

    # Combine arrays into one pandas dataframe
    df = pd.DataFrame({"2": list(clusters2), "3": list(clusters3), "4" : list(clusters4), 
                       "5" : list(clusters5), "6" : list(clusters6)}, 
                      columns=["2", "3", "4", "5", "6"])

    # Create boxplot with pandas dataframe
    boxplot = df.boxplot(column=["2", "3", "4", "5", "6"], figsize=(10, 6))
    plt.ylabel("Final accuracy")
    plt.xlabel("# of Clusters")

    plt.title("Varied Clusters - " + DataSetName + " (K-Means)")
```

# Iris

```python
## Iris data set ##
plotAccVsClusters("Iris", "iris-data/iris_clusters2_accuracy.txt", "iris-data/iris_clusters3_accuracy.txt",
                 "iris-data/iris_clusters4_accuracy.txt", "iris-data/iris_clusters5_accuracy.txt",
                 "iris-data/iris_clusters6_accuracy.txt")
```
