```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import fowlkes_mallows_score
from sklearn import datasets
import numpy as np
```

```python
def fowlkes_mallows_clustering(X, y):
    """Takes in a data set X and target labels y, and returns Fowlkes-Mallows scoreres and plot"""
    #List of Fowlkes-Mallows Scores for different models
    fms_scores = []

    #List of different number of clusters
    N_Clusters = [2,3,4,5,6]
    
    ### 2 clusters ###
    #Building the clustering model
    kmeans2 = KMeans(n_clusters=2)

    #Training the clustering model
    kmeans2.fit(X)

    #Storing the predicted Clustering labels
    labels2 = kmeans2.predict(X)

    #Evaluating the performance
    fms_scores.append(fowlkes_mallows_score(y,labels2))
    
    ### 3 Clusters ###
    #Building the clustering model
    kmeans3 = KMeans(n_clusters=3)

    #Training the clustering model
    kmeans3.fit(X)

    #Storing the predicted Clustering labels
    labels3 = kmeans3.predict(X)

    #Evaluating the performance
    fms_scores.append(fowlkes_mallows_score(y,labels3))

    ### 4 Clusters ###
    #Building the clustering model
    kmeans4 = KMeans(n_clusters=4)

    #Training the clustering model
    kmeans4.fit(X)

    #Storing the predicted Clustering labels
    labels4 = kmeans4.predict(X)

    #Evaluating the performance
    fms_scores.append(fowlkes_mallows_score(y,labels4))

    ### 5 Clusters ###
    #Building the clustering model
    kmeans5 = KMeans(n_clusters=5)

    #Training the clustering model
    kmeans5.fit(X)

    #Storing the predicted Clustering labels
    labels5 = kmeans5.predict(X)

    #Evaluating the performance
    fms_scores.append(fowlkes_mallows_score(y,labels5))

    
    ### 6 Clusters ###
    #Building the clustering model
    kmeans6 = KMeans(n_clusters=6)

    #Training the clustering model
    kmeans6.fit(X)

    #Storing the predicted Clustering labels
    labels6 = kmeans6.predict(X)

    #Evaluating the performance
    fms_scores.append(fowlkes_mallows_score(y,labels6))
    
    print(fms_scores)
    #Plotting a Bar Graph to compare the models
    plt.figure(figsize=(8, 6), dpi=80)
    plt.scatter(N_Clusters,fms_scores)
    plt.xlabel('Number of Clusters')
    plt.ylabel('Fowlkes Mallows Score')
    plt.title('Comparison of different Clustering Models')
    plt.show()
```

# Iris

```python
# Load IRIS dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

fowlkes_mallows_clustering(X,y)
```

```python
# Load Breast Tissue data set
breast_tissue = np.loadtxt(r'../datasets/breast-tissue.txt')

X = breast_tissue[:, :-1]
y = breast_tissue[:, -1]
fowlkes_mallows_clustering(X, y)
```
