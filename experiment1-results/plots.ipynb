```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
def plot_performance(fileName):
    # Load performance data as numpy array
    data_array = np.loadtxt(fileName, dtype=float)
    data_array_size = np.size(data_array)

    # Plot model performance
    x_axis = [i for i in range(data_array_size)]  # set x-axis to array indices
    plt.figure(figsize=(8, 6), dpi=80)
    plt.scatter(x_axis, data_array, s=5)
    plt.xlabel("Data point")
    plt.ylabel("Final accuracy")
    plt.ylim([0,1])
    plt.yticks(np.arange(0, 1, .1))
```

## Neural Network w/ Heart Disease Data

```python
plot_performance('heart_accuracy.txt')
plt.title("Neural net performance on Heart disease data")
```

## Neural Network w/ Wine Data 

```python
plot_performance('wine_accuracy.txt')
plt.title("Neural net performance on Wine data")
```

## K-Means w/ Iris Data 

```python
plot_performance('iris_accuracy.txt')
plt.title("K-means performance on Iris data")
```

## K-Means w/ Cancer Data 

```python
plot_performance('cancer_accuracy.txt')
plt.title("K-means performance on Cancer data")
```
